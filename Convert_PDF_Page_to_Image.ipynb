```python
def Read_table(doc,page_no,cordinates):
  table_area=','.join([str(cordinates[0]),str(cordinates[1]),str(cordinates[2]),str(cordinates[3])])
  tables1 = camelot.read_pdf(doc, pages=str(page_no),table_areas = [table_area])
  return tables1

```
